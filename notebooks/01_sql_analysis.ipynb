```python
import sqlite3
import pandas as pd
from pathlib import Path

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
pd.set_option('display.max_colwidth', 50)

# Connect to DB
DB_PATH = Path.cwd().parent / "data" / "olist.db"
conn = sqlite3.connect(DB_PATH)

def q(sql_or_path):
    """Run SQL — accepts either a SQL string or a path to a .sql file."""
    p = Path(sql_or_path)
    if p.exists():
        sql = p.read_text()
    else:
        sql = sql_or_path
    return pd.read_sql(sql, conn)

print("Connected to:", DB_PATH)
print("Tables:", pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table' ORDER BY name", conn
)['name'].tolist())
```
